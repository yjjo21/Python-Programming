```python
import matplotlib.pyplot as plt
```

subplot 과 subplots는 사용법이 다르다. **subplot**은 일일히 다 설정하고, 하나의 axes만을 나타내지만 **subplots**는 figure,axes를 함께 생성한다.

### 1. subplot

```python
plt.subplot(2,1,1)
plt.subplot(2,1,2)
plt.show()
```

### 2. subplots

```python
fig, ax = plt.subplots()
```
