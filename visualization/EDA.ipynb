```python
DF = load_penguins()
DF.dropna(inplace = True)
print(DF.shape)
DF.head()
```

# 인덱스 순의 시각화

```python
fig, ax = plt.subplots(figsize = (15, 4))
sns.scatterplot(data = DF,
                x = DF.index,
                y = DF['bill_length_mm'],
                s = 5,
                color = 'black',
                edgecolor = 'None')
```

### matplot 을 이용하는 방법도 있다. 

```python
fig, ax = plt.subplots(figsize = (15, 4))
plt.scatter(x = DF.index,
            y = DF['year'],
            s = 5,
            color = 'black',
            edgecolor = 'None')
start, end = ax.get_xlim()
print(start, end)
#ax.xaxis.set_ticks(np.arange(start, end, 50))
ax.set_xticks([i for num,i in enumerate(ax.get_xticks().tolist()) if num%2 ==0])
plt.xticks(fontsize =14)
plt.yticks(fontsize = 14)
plt.show()
```
